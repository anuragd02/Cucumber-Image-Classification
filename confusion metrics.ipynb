```python
import matplotlib.pyplot as plt
import numpy as np

# Define the scale factor for the image size
scale_factor = 0.8  # You can change this value to adjust the image size

# Define the new confusion matrix
confusion_matrix = np.array([[70, 0],
                              [14, 147]])

# Define axis tick labels
tick_labels = ['Healthy', 'Unhealthy']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8 * scale_factor, 6 * scale_factor))

# Display the confusion matrix as a heatmap
cax = ax.matshow(confusion_matrix, cmap='Blues')

# Add color bar
plt.colorbar(cax)

# Set axis labels
ax.set_xlabel('Predicted label', fontsize=12)
ax.set_ylabel('True label', fontsize=12)

# Set tick labels
ax.set_xticks(np.arange(len(tick_labels)))
ax.set_yticks(np.arange(len(tick_labels)))
ax.set_xticklabels(tick_labels, fontsize=10)
ax.set_yticklabels(tick_labels, fontsize=10)

# Annotate each cell with the numeric value
for i in range(confusion_matrix.shape[0]):
    for j in range(confusion_matrix.shape[1]):
        ax.text(j, i, confusion_matrix[i, j], ha='center', va='center', 
                color='white' if confusion_matrix[i, j] > confusion_matrix.max() / 2 else 'black')

# Add a title to the plot
plt.title(r'$\bf{Convolutional\ Neural\ Network}$', fontsize=14, pad=20)

# Adjust the x-axis labels' position to be closer to the table
ax.xaxis.set_label_position('bottom')
ax.xaxis.tick_bottom()
plt.xticks(ticks=np.arange(len(tick_labels)), labels=tick_labels, fontsize=10, rotation=0)
plt.tight_layout()

plt.show()
```
